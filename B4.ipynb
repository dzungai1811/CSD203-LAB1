```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def addToHead(self, x):
        new_node = Node(x)
        if not self.head:
            new_node.next = new_node
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.tail.next = new_node
            self.head = new_node

    def addToTail(self, x):
        new_node = Node(x)
        if not self.tail:
            new_node.next = new_node
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.tail.next = new_node
            self.tail = new_node

    def addAfter(self, p, x):
        current = self.head
        while current and current.data != p:
            current = current.next
            if current == self.head:
                break
        if current:
            new_node = Node(x)
            new_node.next = current.next
            current.next = new_node
            if current == self.tail:
                self.tail = new_node

    def traverse(self):
        if not self.head:
            return
        current = self.head
        while True:
            print(current.data, end=" -> ")
            current = current.next
            if current == self.head:
                break
        print("Head")

    def deleteFromHead(self):
        if not self.head:
            return None
        data = self.head.data
        if self.head == self.tail:
            self.head = None
            self.tail = None
        else:
            self.head = self.head.next
            self.tail.next = self.head
        return data

    def deleteFromTail(self):
        if not self.tail:
            return None
        data = self.tail.data
        if self.head == self.tail:
            self.head = None
            self.tail = None
        else:
            current = self.head
            while current.next != self.tail:
                current = current.next
            current.next = self.head
            self.tail = current
        return data

    def deleteAfter(self, p):
        current = self.head
        while current and current.data != p:
            current = current.next
            if current == self.head:
                break
        if current and current.next != self.head:
            data = current.next.data
            current.next = current.next.next
            if current.next == self.head:
                self.tail = current
            return data
        return None

    def delete(self, x):
        if not self.head:
            return None
        if self.head.data == x:
            return self.deleteFromHead()
        current = self.head
        while current.next and current.next.data != x:
            current = current.next
            if current == self.head:
                break
        if current.next and current.next != self.head:
            data = current.next.data
            current.next = current.next.next
            if current.next == self.head:
                self.tail = current
            return data
        return None
cll = CircularLinkedList()
cll.addToHead(1)
cll.addToTail(2)
cll.addAfter(1, 3)
cll.traverse()
print("Deleted from Head:", cll.deleteFromHead())
cll.traverse()
print("Deleted from Tail:", cll.deleteFromTail())
cll.traverse()
cll.addAfter(2, 4)
cll.traverse()
print("Deleted node with value 3:", cll.delete(3))
cll.traverse()
```
